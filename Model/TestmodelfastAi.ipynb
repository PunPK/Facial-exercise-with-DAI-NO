```python
def conv2(ni, nf): return ConvLayer(ni, nf, stride=2)

class ResBlock(Module):
  def __init__(self, nf):
    self.conv1 = ConvLayer(nf, nf)
    self.conv2 = ConvLayer(nf, nf)
  
  def forward(self, x): return x + self.conv2(self.conv1(x))

def conv_and_res(ni, nf): return nn.Sequential(conv2(ni, nf), ResBlock(nf))
```

```python
import tensorflow as tf
from tensorflow import keras

loaded_model = keras.models.load_model('yawn_detection.h5')
```

```python
import cv2
import numpy as np

# อ่านและตรวจสอบภาพ
image = cv2.imread("_2.jpg", 0)
if image is not None:
    # ปรับขนาดภาพ
    resized_image = cv2.resize(image, (64, 64))
    input_data = np.expand_dims(resized_image, axis=-1)
    input_data = np.expand_dims(input_data, axis=0)

    # ทำนายภาพโดยใช้โมเดล
    predictions = loaded_model.predict(input_data)
    class_names = ['close eye', 'open eye']
    predicted_class = np.argmax(predictions)
    predicted_label = class_names[predicted_class]
    print(predicted_label)
else:
    print("Failed to load the image.")
```
